```python
import mule
import numpy as np
import xarray as xr
import os
```

```python
data = xr.open_dataset('/g/data/p66/txz599/ACCESS-ESM_tools/ancil/tmp/co2_new.nc')
data_ts = xr.open_dataset('/g/data/p66/txz599/ACCESS-ESM_tools/ancil/orig_co2.nc')
```

```python
# 10 GtC/yr constant emissions, globally homogenous
# 10 GtC/yr = 36.7 GtCO2/yr = 36.7e12 kgCO2/yr
# yr to s : 60*60*24*365
# emissions field required in kgCO2/m2/s

yr2s = 60*60*24*365.25 # s
earea = 510e12 # m2

# calculation for 10 PgC/yr
co2emissions = 10 * 3.67 * 1e12 / yr2s / earea

# calculation for 12 PgC/yr
#co2emissions = 12 * 3.67 * 1e12 / yr2s / earea

```

```python
data.co2emiss[:,:,:] = co2emissions
```

```python
co2emissions_global = np.zeros(201)
co2emissions_lin = np.zeros(201)

co2_init = 10. # PgC/yr
co2_end = -10.

co2emissions_global[0] = co2_init
for i in range(100):
    co2emissions_global[i+1] = co2emissions_global[i] - 0.2 

# adjust centre value to be exactly 0
co2emissions_global[50] = 0.

# constant emissions -10 after 100 years
co2emissions_global[101:201] = co2_end
```

```python
co2emissions_global
```

```python
#convert to kgCO2/m2/s

co2_emit_lin = np.zeros(201)
for i in range(201):
    #print(i)
    co2emissions_lin[i] = co2emissions_global[i] * 3.67 * 1e12 / yr2s / earea
    
```

```python
co2emissions_lin
```

```python
new_co2_field = data_ts.co2emiss[0:2412,:,:]
#new_co2_field = np.zeros(1200)
for i in range(201):
    new_co2_field[12*i:12*(i+1),:,:] = co2emissions_lin[i]
```
